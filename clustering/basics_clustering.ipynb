# Import dependencies

```python
import os
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans, DBSCAN

import matplotlib.pyplot as plt
import seaborn as sns

from scipy.cluster import hierarchy
from scipy.spatial.distance import pdist
```

# Data Prep

```python
x1 = np.random.normal(25, 2, 100)
y1 = np.random.normal(25, 2, 100)

x2 = np.random.normal(55, 2, 100)
y2 = np.random.normal(60, 3, 100)

x3 = np.random.normal(40, 2, 100)
y3 = np.random.normal(5, 2, 100)

x4 = np.random.normal(50, 3, 100)
y4 = np.random.normal(30, 3, 100)
```

```python
x = np.concatenate((x1,x2,x3,x4), axis=0)
y = np.concatenate((y1,y2,y3,y4), axis=0)

df = pd.DataFrame({'x':x, 'y':y})
```

```python
sns.scatterplot(x='x', y='y', data=df)
plt.show()
```

# K-Means Clustering

```python
kmeans = KMeans(n_clusters=4).fit(df)
```

```python
sns.scatterplot(x='x', y='y', data=df, hue=kmeans.labels_, palette='viridis')
plt.show()
```

# DBSCAN Clustering

```python
db = DBSCAN(eps=5, min_samples=20).fit(df)
```

```python
sns.scatterplot(x='x', y='y', data=df, hue=db.labels_, palette='viridis')
plt.show()
```

# Hierarchical Clustering

```python
dist = pdist(df)
```

```python
z=hierarchy.linkage(dist,'single')
hierarchy.dendrogram(z, color_threshold=3)

plt.show()
```

```python
sns.scatterplot(x='x', y='y', data=df, hue=hierarchy.fcluster(z, 3, criterion='distance'), palette='viridis')

plt.show()
```
